```python
import yaml
```

```python
from glob import glob

img_list = glob('/content/drive/MyDrive/project_BigData/2021.10.27/train/images/*.jpg')

print(len(img_list))
```

```python
from sklearn.model_selection import train_test_split
train_img_list, val_img_list = train_test_split(img_list, test_size=0.1, random_state = 2000)
print(len(train_img_list), len(val_img_list))
```

```python
with open('/content/drive/MyDrive/project_BigData/2021.10.27/train.txt', 'w') as f:
  f.write('\n'.join(train_img_list)+'\n')
with open('/content/drive/MyDrive/project_BigData/2021.10.27/val.txt', 'w') as f:
  f.write('\n'.join(val_img_list)+'\n')  
```

```python
with open('/content/drive/MyDrive/project_BigData/2021.10.27/data.yaml','r') as f:
  data = yaml.load(f)
data  
```

```python
data['train'] = '/content/drive/MyDrive/project_BigData/2021.10.27/train.txt'
data['val'] = '/content/drive/MyDrive/project_BigData/2021.10.27/val.txt'

with open('/content/drive/MyDrive/project_BigData/2021.10.27/data.yaml','w') as f:
  yaml.dump(data, f)
print(data)  
```
